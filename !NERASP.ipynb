```python
import pandas as pd
import numpy as np
```

```python
#заносим фрейм в переменную
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/!NERASP/ZHILIE_SVOD22.csv', encoding='utf-8', sep='\t')
```

```python
#define a function to clean and convert a column to float
def clean_and_convert(column):
    # Replace commas with periods for decimal places
    column = column.str.replace(',', '.')

    # Remove any non-numeric characters except periods
    column = column.str.replace('[^0-9.]', '', regex=True)

    # Convert the column to float, setting errors='coerce' to handle empty strings and non-convertible values
    column = pd.to_numeric(column, errors='coerce')

    return column
```

```python
# Apply the function to the specified columns
df['PLOSHAD'] = clean_and_convert(df['PLOSHAD'])
df['NAGRUZKA'] = clean_and_convert(df['NAGRUZKA'])
df['PNR'] = clean_and_convert(df['PNR'])
df['TARIF_TEPLO'] = clean_and_convert(df['TARIF_TEPLO'])
```

```python
df['DATE_START'] = pd.to_datetime(df['DATE_START'], dayfirst=True)
df['DATE_OUT'] = pd.to_datetime(df['DATE_OUT'], dayfirst=True)
```

```python
def RASCHET_SODERG(df):
  if df["DATE_START"] == pd.Timestamp("2022-01-01"):
    if df["DAYS"] <= 31:
      return round(df["PLOSHAD"] * df["PNR"] / 31 * df["DAYS"], 2)
    elif 32 <= df["DAYS"] <= 58:
      return round(df["PLOSHAD"] * df["PNR"] + df["PLOSHAD"] * df["PNR"] / 28 * (df["DAYS"] - 31), 2)
    elif 59 <= df["DAYS"] <= 91:
      return round(df["PLOSHAD"] * df["PNR"] * 2 + df["PLOSHAD"] * df["PNR"] / 31 * (df["DAYS"] - 58), 2)
  elif df["DATE_START"] == pd.Timestamp("2022-04-01"):
    if df["DAYS"] <= 30:
      return round(df["PLOSHAD"] * df["PNR"] / 30 * df["DAYS"], 2)
    elif 31 <= df["DAYS"] <= 61:
      return round(df["PLOSHAD"] * df["PNR"] + df["PLOSHAD"] * df["PNR"] / 31 * (df["DAYS"] - 30), 2)
    elif 62 <= df["DAYS"] <= 91:
      return round(df["PLOSHAD"] * df["PNR"] * 2 + df["PLOSHAD"] * df["PNR"] / 30 * (df["DAYS"] - 61), 2)
  elif df["DATE_START"] == pd.Timestamp("2022-07-01"):
    if df["DAYS"] <= 31:
      return round(df["PLOSHAD"] * df["PNR"] / 31 * df["DAYS"], 2)
    elif 32 <= df["DAYS"] <= 62:
      return round(df["PLOSHAD"] * df["PNR"] + df["PLOSHAD"] * df["PNR"] / 31 * (df["DAYS"] - 31), 2)
    elif 63 <= df["DAYS"] <= 92:
      return round(df["PLOSHAD"] * df["PNR"] * 2 + df["PLOSHAD"] * df["PNR"] / 30 * (df["DAYS"] - 62), 2)
  elif df["DATE_START"] == "2022-10-01":
    if df["DAYS"] <= 31:
      return round(df["PLOSHAD"] * df["PNR"] / 31 * df["DAYS"], 2)
    elif 32 <= df["DAYS"] <= 61:
      return round(df["PLOSHAD"] * df["PNR"] + df["PLOSHAD"] * df["PNR"] / 30 * (df["DAYS"] - 31), 2)
    elif 62 <= df["DAYS"] <= 92:
      return round(df["PLOSHAD"] * df["PNR"] * 2 + df["PLOSHAD"] * df["PNR"] / 31 * (df["DAYS"] - 61), 2)
  else:
        return 'error!!!!'
```

```python
def RASCHET_TEPLO(df):
  if df["DATE_START"] == pd.Timestamp("2022-01-01"):
    if df["DAYS"] <= 31:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] / 31 * df["DAYS"], 2)
    elif 32 <= df["DAYS"] <= 58:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 28 * (df["DAYS"] - 31), 2)
    elif 59 <= df["DAYS"] <= 91:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] * 2 + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 31 * (df["DAYS"] - 58), 2)
  elif df["DATE_START"] == pd.Timestamp("2022-04-01"):
    if df["DAYS"] <= 30:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] / 30 * df["DAYS"], 2)
    elif 31 <= df["DAYS"] <= 61:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 31 * (df["DAYS"] - 30), 2)
    elif 62 <= df["DAYS"] <= 91:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] * 2 + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 30 * (df["DAYS"] - 61), 2)
  elif df["DATE_START"] == pd.Timestamp("2022-07-01"):
    if df["DAYS"] <= 31:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] / 31 * df["DAYS"], 2)
    elif 32 <= df["DAYS"] <= 62:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 31 * (df["DAYS"] - 31), 2)
    elif 63 <= df["DAYS"] <= 92:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] * 2 + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 30 * (df["DAYS"] - 62), 2)
  elif df["DATE_START"] == "2022-10-01":
    if df["DAYS"] <= 31:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] / 31 * df["DAYS"], 2)
    elif 32 <= df["DAYS"] <= 61:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 30 * (df["DAYS"] - 31), 2)
    elif 62 <= df["DAYS"] <= 92:
      return round(df["NAGRUZKA"] * df["TARIF_TEPLO"] * 2 + df["NAGRUZKA"] * df["TARIF_TEPLO"] / 31 * (df["DAYS"] - 61), 2)
  else:
        return 'error!!!!'
```

```python
# Применяем пользовательскую функцию к каждой строке датафрейма
df["SODERG"] = df.apply(RASCHET_SODERG, axis=1)
df["TEPLO"] = df.apply(RASCHET_TEPLO, axis=1)
df['SUMMA'] = df["SODERG"] + df["TEPLO"]
```
